```python
newdf = df.drop(columns=['ID','engine_power','lat','lon','price'])
print(newdf.to_string())
```

```python
newdf1 = pd.get_dummies(newdf,dtype=int)
print(newdf1.to_string())
```

```python
X1 = newdf1
Y1 = df['price']
reg = LinearRegression()
reg.fit(X1,Y1)
```

```python
ans = reg.predict([[2190,70000,1,0,0,1],[2190,70000,1,1,0,0],[2190,70000,1,0,1,0]])
print(ans)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X1,Y1,test_size=0.33)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
finalans = LinearRegression()
finalans.fit(X_train,Y_train)
ans1 = finalans.predict(X_test)
print(ans1)
```

```python
import seaborn as sns
import matplotlib as plt
result = pd.DataFrame(columns=['KM','Price'])
result['KM'] = X_test['km']
result['Price'] = ans1
sns.lineplot(x='KM',y='Price',data=result)
```
